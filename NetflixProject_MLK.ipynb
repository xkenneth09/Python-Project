```python
plt.style.use('seaborn-v0_8-whitegrid')
ax = sns.countplot(x="Type", data=netflix)
```

```python
categories = set([s for Genre in netflix.Genre.unique() for s in Genre.split(",")])
```

```python
for cat in categories:
    netflix[cat] = netflix.Genre.transform(lambda s: int(cat in s))
df = netflix[['Director','Genre','Duration'] + list(categories)]
df.sample(5)
```

```python
df = pd.melt(df,
             id_vars=['Genre'],
             value_vars = list(categories),
             var_name = 'Category',
             value_name = 'Count')
df = df.loc[df.Count>0]
top_categories = df.groupby('Category').aggregate(sum).sort_values('Count', ascending=False).index
howmany=10
```

```python
p = sns.countplot(data=df, x = 'Category')
```

```python
n_movie = netflix[netflix['Type'] == 'Movie'].shape[0]
n_tv_show = netflix[netflix['Type'] == 'TV Show'].shape[0]

data = {'Type': ['Movie', 'TV Show'], 'count': [n_movie, n_tv_show]}
df_bar_data = pd.DataFrame(data)

sns.barplot(x='Type', y='count', data=df_bar_data)

plt.xlabel('Content Type')
plt.ylabel('Number of Titles')
plt.title('Distribution of Content Types on Netflix')
plt.show()
```

```python
netflix['year'] = netflix ['Date_added'].dt.year
yearly_content_count = (
    netflix.groupby(['year', 'Type'])
    .size()
    .to_frame(name='count')
    .reset_index()
)

sns.lineplot(x='year', y='count', hue='Type', data=yearly_content_count)
plt.title('Evolution of Content Over Time on Netflix')
plt.xlabel('Year')
plt.ylabel('Number of Titles Added')
plt.legend(title='Content Type')
plt.xticks(rotation=45)  
plt.tight_layout() 
plt.show()
```
